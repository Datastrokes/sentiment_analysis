Let us do tokenization

```python
import spacy
from nltk.stem import PorterStemmer
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stemming=PorterStemmer()
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
df_list=[]
final=[]
for document in df['review']:
    doc=nlp(document)
    for token in doc:
      if token.text not in stopwords.words('english'):
        df_list.append(stemming.stem(token.text))

    string=' '.join(df_list)
    final.append(string)
    df_list.clear()
```
